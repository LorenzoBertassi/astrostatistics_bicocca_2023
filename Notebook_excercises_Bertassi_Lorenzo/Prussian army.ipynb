```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.stats import poisson
```

```python
#loading data

x=np.array([0,1,2,3,4])
y=np.array([109,65,22,3,1])

#normalizing data

p=y/np.sum(y) 


#computing the expected value from a Poisson distribution

#trying to fit with a poisson distribution with different mu

for m in [0.1,0.3,0.5,0.7,0.9,1]:
    dist=poisson(np.mean(m))
    #computing probabilities
    
    prob=dist.pmf(x)
    
    #plotting for each mu
    
    plt.plot(x, prob,alpha=0.5,label='poisson distribution with $\mu$='+str(m))

#plotting data

plt.scatter(x,p,alpha=0.75,color='black',label='data')

plt.xlabel('number of deaths')
plt.ylabel('probability')

plt.legend()
plt.show()

#finding best mu given data

mu=np.average(x,weights=p)
print('best mean',"{:.2f}".format(mu))
#computing probabilities for this mu

dist=poisson(mu)

#plotting

plt.plot(x, dist.pmf(x),label='poisson distribution with $\mu$='+str("{:.2f}".format(mu)))
plt.scatter(x,p,color='black',alpha=0.5,label='data')

plt.xlabel('number of deaths')
plt.ylabel('probability')

plt.legend()
plt.show()
```
