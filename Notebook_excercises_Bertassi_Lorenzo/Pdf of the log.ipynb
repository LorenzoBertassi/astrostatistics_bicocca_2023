```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sampling the uniform distribution and computing the log_10 

x=np.random.uniform(low=0.1, high=10, size=100000)
y=np.log10(x)

#creating a grid and computing the probability to check if the sampling is correct

x_grid=np.linspace(0.1,10,100000)
p_x=np.ones(len(x_grid))/(10-0.1)
```

```python
#plotting the sampled data and checking with the theoretical one

plt.hist(x,bins=100,density=True,alpha=0.5,label='sampled data');
plt.plot(x_grid,p_x,color='red',label='expected pdf')
plt.xlabel('x')
plt.legend()
plt.show()

#plotting the distribution of y

plt.hist(y,bins=100,density=True,alpha=0.5,label='sampled data');

#computing the expected one for a check applying the tranformation rule:
#p(y)=|dx/dy|p(x)

y_grid=np.linspace(np.min(y),np.max(y),100000)
p_y=p_y=10**y_grid*np.log(10)*1/(10-0.1)

#plotting the expected distribution

plt.plot(y_grid,p_y,color='red',label='expected pdf')
plt.xlabel('$log_{10}(x)$')
plt.legend()
plt.show()
```

```python
#computing the log of the mean of x and the mean of y

log_mean_x=np.log10(np.mean(x))
mean_y=np.mean(y)

#computing the log of the median of x and the median of y

log_median_x=np.log10(np.median(x))
median_y=np.median(y)

#printing the results 
print("We expect the median to not change since a monotonic tranformation doesn't change the order of the samples.")
print('\nmeans',log_mean_x,mean_y)
print('medians',log_median_x,median_y)
print('\ncheck that the median are the same: ', np.isclose(log_median_x,median_y))
```
