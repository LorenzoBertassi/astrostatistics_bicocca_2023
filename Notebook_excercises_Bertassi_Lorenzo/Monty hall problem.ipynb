```python
import numpy as np
import random
import time
```

Let's call True the event that the prize is in the door and False the event in which the prize isn't there.
When we are considering all the three doors, let's choose a door and call it door 1. The probability of finding the prize in 1 is given by $\frac{1}{3}$. Let's now assume that a door is open, let's say door 3, and there is no prize inside.

We can compute the probability of finding 1=True given 3=False using bayes theorem:
$p(1=True|3=False)=\frac{p(3=False|1=True)p(1=True)}{p(3=False|1=True)p(1=True)+p(3=False|1=False)p(1=False)}$

Substuting the values for the known probabilities we get:
$p(1=True|3=False)=\frac{\frac{2}{3}\frac{1}{3}}{\frac{2}{3}\frac{1}{3}+\frac{2}{3}\frac{2}{3}}=\frac{1}{3}$


We understand that to win the prize it's better to change door.

```python
#Game with 3 doors

#initializing counts to zero

#keeper
k=0

#changer
c=0

#newcomer
n=0

#number of games
game=0


random.seed(int(time.time()))

#Games

for g in range (0,10000):
    doors=[1,2,3]
    
    #choosing the door with the car
    
    car=random.randint(0,2)
    doors[car]='car'
    
    i=0
    for i in range (0,len(doors)):
        
        #putting goats in the other two doors
        
        if (i != car):
            doors[i]='goat'
            
    #making the choice
    
    choice=random.randint(0,2)
    
    #opening one door
    
    p=random.randint(0,2)
    while(p==car or p==choice):
        p=random.randint(0,2)
    doors.pop(p)
    if(p<choice):
        choice=choice-1
    
    #Upgrading winners
    
    if(doors[choice]=='car'):
        k+=1
    else:
        c+=1
        
    #making the newcomer choose
    
    new=random.randint(0,1)
    if(doors[new]=='car'):
        n+=1
    game+=1
    
#printing the results

print('keeper:  ',k/game,'\nchanger: ',c/game,'\nnewcomer:',n/game)
```

We can easyly generalize the problem assuming that the probability of having 1=True (we are again assuming that we choose door 1) is given by $\frac{1}{N}$. Let's assume that N-2 doors in which surely there is no prize are opened. 

We can again compute the probability using Bayes theorem. Let's call E the set of N-2 doors in which there is no prize then:
$p(1=True|E=False)=\frac{p(E=False|1=True)p(1=True)}{p(E=False|1=True)p(1=True)+p(E=False|1=False)p(E=False)}$

Substuting the values for the known probabilities we get:
$p(1=True|E=False)=\frac{\frac{N-1}{N}\frac{1}{N}}{\frac{N-1}{N}\frac{1}{N}+\frac{N-1}{N}\frac{N-1}{N}}=\frac{1}{N}$


```python
#game with n doors (for 100 doors it may take a while)

#initializing counts
k=0
c=0
n=0
game=0

random.seed(int(time.time()))

#asking the number of doors

number=int(input('Insert the number of doors: '))

#games

while(game<1000000):
    doors=[]
    final_doors=[]
    
    #creating doors
    for a in range (0,number):
        doors.append(1)
    
    #putting the car behind one door
    
    car=random.randint(0,len(doors)-1)
    doors[car]='car'
    
    #putting goats behind the other doors
    
    i=0
    for i in range (0,len(doors)-1):
        if (i != car):
            doors[i]='goat'
            
    #choosing the door
    
    choice=random.randint(0,len(doors)-1)
    if (choice !=car):
        final_doors.append(doors[choice])
        final_doors.append(doors[car])
    if (choice==car):
        i=random.randint(0,len(doors)-1)
        while(i==car):
            i=random.randint(0,len(doors)-1)
        final_doors.append(doors[choice])
        final_doors.append(doors[i])
        
    #upgrading winners
    
    if(final_doors[0]=='car'):
        k+=1
    else:
        c+=1
    
    #making the choice of the newcomer
    
    new=random.randint(0,1)
    if(final_doors[new]=='car'):
        n+=1
    game+=1

#printing results
print('keeper:  ',k/game,'\nchanger: ',c/game,'\nnewcomer:',n/game)
```
